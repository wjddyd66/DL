```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
#sigmoid 함수 선언
def sigmoid(x):
    return 1/(1+np.exp(-x))
#sigmoid 미분 함수 선언
def sigmoid_(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
#Sigmoid 그래프로서 표현하기
x = np.arange(-7.0,7.0,0.1)
y = sigmoid(x)

plt.plot(x,y)
plt.plot([0,0],[1.0,0.0],':')
plt.ylim(-0.1,1.1)
plt.title('Sigmoid Function')
plt.show()
```

```python
#Sigmoid 미분그래프로서 표현하기
x = np.arange(-7.0,7.0,0.1)
y = sigmoid_(x)

plt.plot(x,y)
plt.plot([0,0],[0.5,0.0],':')
plt.ylim(-0.1,0.5)
plt.title('Sigmoid Differential Function')
plt.show()
```

```python
#tanh 함수 선언
def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
#tanh 미분 함수 선언
def tanh_(x):
    return 1-(tanh(x)*tanh(x))
```

```python
#tanh 그래프로서 표현하기
x = np.arange(-7.0,7.0,0.1)
y = tanh(x)

plt.plot(x,y)
plt.plot([0,0],[1.0,-1.0],':')
plt.ylim(-1.1,1.1)
plt.title('tanh Function')
plt.show()
```

```python
#tanh 미분그래프로서 표현하기
x = np.arange(-7.0,7.0,0.1)
y = tanh_(x)

plt.plot(x,y)
plt.plot([0,0],[2.0,0.0],':')
plt.ylim(-0.1,2.0)
plt.title('tanh Differential Function')
plt.show()
```

```python
#ReLU 함수 선언
def relu(x):
    return np.maximum(0,x)
#ReLU 미분 함수 선언
def relu_(x):
    result =[]
    for i in x:
        if(i>=0):
            result.append(1)
        else:
            result.append(0)
    return result
```

```python
#ReLU 그래프로서 표현하기
x = np.arange(-7.0,7.0,0.1)
y = relu(x)

plt.plot(x,y)
plt.plot([0,0],[8.0,-0.5],':')
plt.ylim(-0.3,8.1)
plt.title('ReLU Function')
plt.show()
```

```python
#ReLU 미분그래프로서 표현하기
x = np.arange(-7.0,7.0,0.1)
y = relu_(x)

plt.plot(x,y)
plt.plot([0,0],[1.5,-0.5],':')
plt.ylim(-0.5,1.5)
plt.title('ReLU Differential Function')
plt.show()
```

```python
def softmax(a):
    max = np.max(a)
    value = np.exp(a)
    sum = np.sum(value)
    result = value/sum
    return result
```

```python
#Softmax그래프로서 표현하기
x = np.array([100,200,300])
y = softmax(x)

plt.plot(x,y)
plt.plot([0,0],[1.1,-0.1],':')
plt.ylim(-0.1,1.1)
plt.title('Softmax Function')
plt.show()
```
